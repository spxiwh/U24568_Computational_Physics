## Exercise 1

Write a generator that returns $x !$ (x-factorial) for x between 1 and 40.

```python
def factorial_generator():
    i = 1
    fac = 1
    while i <= 40:
        yield fac
        fac *= i
        i += 1

gen = factorial_generator()

# to confirm that this is correct, we compare to the factorial function in the math module,
# utilising the enumerate() function to get the index of the current factorial
from math import factorial
for i, curr_fac in enumerate(gen):
    print (curr_fac)
    assert curr_fac == factorial(i)

```

## Exercise 2

Write a generator that loops over values of $x^2$ while $x^2$ is smaller than 10000.

```python
def square_generator():
    i = 1
    while i ** 2 <= 10000:
        yield i ** 2
        i += 1

gen = square_generator()

for curr_square in gen:
    print (curr_square)
```
