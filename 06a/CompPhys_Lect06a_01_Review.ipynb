## Making a basic plot

You are measuring the voltage as a function of time from mains electrics in the UK. How should we plot this?

```python
mains_frequency = 50 * (2*np.pi)

sample_time = np.linspace(0, 0.04, 101)
voltage_0 = 230 * np.sin(mains_frequency * sample_time)

plt.plot(sample_time, voltage_0)
```

Basic additions:
 - colours. See [here](https://html-color-codes.info/color-names/) for a full list of colours.
 - markers and linestyles, see [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) for a list
 - a grid using `grid()`
 - a legend using `legend()`
 - axis labels

Note that in the coursework, if your plots are not appropriately labelled, then you will not be able to gain full marks.

```python
voltage_1 = 230 * np.sin(mains_frequency * sample_time + 2 * np.pi / 3)
voltage_2 = 230 * np.sin(mains_frequency * sample_time + 4 * np.pi / 3)

plt.plot(sample_time, voltage_1, c="Goldenrod", linestyle='--', label="Supply 1")
plt.plot(sample_time, voltage_2, c="Fuchsia", linestyle=':', label="Supply 2")

plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.grid()
plt.legend()
```

## Histograms

We can do two things for generating histograms, mainly depending on whether we want to use the output of the histogram for something else.
* Use matplotlib to generate the histogram
* Generate the histogram using numpy, and then plot that output

```python
# Here we will generate some random numbers for use in the following examples
r = np.random.normal(size=1000)
```

```python
plt.hist(r)
plt.xlabel("Random number")
plt.ylabel("Count")
```

```python

counts, bin_edges = np.histogram(r)
# Numpy's histogram function will give the bin edges, so there will be one more value than the counts in the bins
# so we need to deal with this in the plotting.
# As a result, we use stairs rather than plot

plt.stairs(counts, bin_edges) # Note the reverse of the order compared to plot
plt.xlabel('Random number')
plt.ylabel("Count")
```
