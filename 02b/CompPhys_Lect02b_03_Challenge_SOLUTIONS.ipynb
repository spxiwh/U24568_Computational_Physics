```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('emission_line.txt', header=None, names=['vals'], dtype=np.float64)
```

```python
# Plot the data
import matplotlib.pyplot as plt
import numpy.random as rnd
import scipy.stats as sci_rnd

# bins = np.linspace(dataset['vals'].min(), dataset['vals'].max(), 10)
bins = 10
plt.hist(dataset, density=True);

# Try to fit a Poisson to distribution
poisson_dist = sci_rnd.poisson.rvs(dataset['vals'].mean(), size=10000)
plt.hist(poisson_dist, density=True, alpha=0.5);
#plt.plot(x_axis, sci_rnd.poisson.pdf(x_axis,5,2))
```

```python
plt.hist(dataset, density=True);

x = np.arange(-1, 7, 1)

# poisson distribution data for y-axis
y = sci_rnd.poisson.pmf(x, mu=dataset['vals'].mean())


# plotting the graph
plt.plot(x, y)
```
