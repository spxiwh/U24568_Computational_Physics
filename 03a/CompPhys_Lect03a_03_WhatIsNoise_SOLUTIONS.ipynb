```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rnd
import scipy.stats as stats
```

```python
# Answer here

# White noise is a random signal having equal intensity at different frequencies, giving it a constant power spectral density (power per frequency interval).
# Pink noise is also known as 1‚ÅÑf noise. Here the power spectral density is inversely proportional to the frequency of the signal
# Blue noise's power density increases 10log2 = 3.01 dB per octave with increasing frequency (density proportional to f )
# Grey noise is random white noise subjected to a psychoacoustic equal loudness curve (such as an inverted A-weighting curve) over a given range of frequencies.
```

```python
# Answer here

# White Noise
white_noise = rnd.uniform(-1., 1.,size=1000)

# Plot
x_data = np.arange(0, len(white_noise), 1)
plt.plot(x_data, white_noise, zorder=0)

# Randomly draw from this noise

draw_x = rnd.uniform(0, 1000 ,10000).astype(int)
draw_y = rnd.uniform(-1, 1, 10000)

# Match to within 2 decimal places
accept_x = draw_x[np.round(draw_y, 2) == np.round(white_noise[draw_x], 2)]
accept_y = draw_y[np.round(draw_y, 2) == np.round(white_noise[draw_x], 2)]

# Accepted points shown in red
plt.scatter(accept_x, accept_y, color='red', zorder=1)
```
