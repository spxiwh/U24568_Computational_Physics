## Pythagorean Triples

Find all of the natural numbers $a$, $b$, and $c$ which satisfy $a^2 + b^2 = c^2$. The simplest example is $3^2 + 4^2 = 5^2$. You can simplify things by only considering $b <= a$, since swapping $a$ and $b$ obviously also gives a solution.

How high in $a$ and $b$ can you search in a reasonable time?

```python
# Answer goes here
a = []
b = []
for i in range(0,10000):
  for j in range(0,10000):
    if j <=i:
      c = np.sqrt(i**2 + j**2)
      if c.is_integer() and j!=0:
        a.append(i)
        b.append(j)
```

```python
print('Iterating over 10000 points takes a while so I would not go higher than this')
```

Challenge: Modify your code to store the tuples `(a, b, c)`, then sort them by $c$ before printing them out.

```python
# Answer goes here
all = []
for i in range(0,100):
  for j in range(0,100):
    if j <=i:
      c = np.sqrt(i**2 + j**2)
      if c.is_integer() and j!=0:
        all.append((i,j,int(c))) # List of tuples

# Sort by c and print the tuples
all.sort(key=lambda x:x[2])

# A bit easier to see the list of tuples as the following
for a in all:
  print(a)
```
